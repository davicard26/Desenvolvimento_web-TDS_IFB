```python
# TAREFA 4
#
# Fa√ßa um programa que solicita uma quantidade indefinida de notas de uma prova
# (deve parar quando um n√∫mero negativo for digitado).
# Depois mostra a m√©dia das notas.

notas = []

while True:
    nota = float(input("Digite uma nota (ou n√∫mero negativo para parar): "))
    if nota < 0:
        break
    notas.append(nota)

if len(notas) > 0:
    media = sum(notas) / len(notas)
    print("Notas digitadas:", notas)
    print("M√©dia das notas:", round(media, 2))
else:
    print("Nenhuma nota v√°lida foi digitada.")
```

```python
TAREFA 5
#
# Fa√ßa um programa que solicita uma quantidade indefinida de n√∫meros positivos
# (deve parar quando um n√∫mero negativo for digitado).
# Armazene-os em duas listas:
# - uma para os n√∫meros pares;
# - outra para os n√∫meros √≠mpares.
# A seguir, mostre a porcentagem de pares e √≠mpares digitados.

pares = []
impares = []

while True:
    num = int(input("Digite um n√∫mero positivo (ou n√∫mero negativo para parar): "))
    if num < 0:
        break
    if num % 2 == 0:
        pares.append(num)
    else:
        impares.append(num)

total = len(pares) + len(impares)

if total > 0:
    perc_pares = (len(pares) / total) * 100
    perc_impares = (len(impares) / total) * 100

    print("\nN√∫meros pares:", pares)
    print("N√∫meros √≠mpares:", impares)
    print(f"\nPorcentagem de pares: {perc_pares:.1f}%")
    print(f"Porcentagem de √≠mpares: {perc_impares:.1f}%")
else:
    print("Nenhum n√∫mero positivo foi digitado.")
```

```python
# TAREFA 6

# Fa√ßa um programa que solicita uma quantidade indefinida de n√∫meros positivos
# (deve parar quando um n√∫mero negativo for digitado).
# Armazene todos os n√∫meros digitados em uma lista, sem repeti√ß√£o.

numeros = []

while True:
    num = int(input("Digite um n√∫mero positivo (ou negativo para parar): "))
    if num < 0:
        break
    if num not in numeros:
        numeros.append(num)

print("\nN√∫meros digitados (sem repeti√ß√£o):", numeros)
```

```python
# TAREFA 7

# Escreva um programa que solicita que o usu√°rio digite a altura e o peso de 5 pessoas.
# Verifique se pelo menos duas das pessoas t√™m a mesma altura e mesmo peso.

pessoas = []

for i in range(5):
    print(f"\nPessoa {i + 1}:")
    altura_str = input("Digite a altura (em metros): ")
    peso_str = input("Digite o peso (em kg): ")


    altura = float(altura_str.replace(',', '.'))
    peso = float(peso_str.replace(',', '.'))

    pessoas.append((altura, peso))

# Verifica duplicatas
duplicado = False
for i in range(len(pessoas)):
    for j in range(i + 1, len(pessoas)):
        if pessoas[i] == pessoas[j]:
            duplicado = True
            break

if duplicado:
    print("\n‚úÖ Pelo menos duas pessoas t√™m a mesma altura e peso.")
else:
    print("\n‚ùå Nenhuma pessoa tem altura e peso iguais.")
```

```python

# üë¥ TAREFA 8

# Fa√ßa um programa que solicita que o usu√°rio digite o nome e a idade de diversas pessoas
# (o programa deve parar quando um nome vazio for informado).
# Retorne o nome da pessoa mais velha.

pessoas = []

while True:
    nome = input("Digite o nome da pessoa (ou pressione Enter para parar): ").strip()
    if nome == "":
        break
    idade = int(input(f"Digite a idade de {nome}: "))
    pessoas.append((nome, idade))

if pessoas:
    mais_velha = max(pessoas, key=lambda p: p[1])
    print(f"\nA pessoa mais velha √© {mais_velha[0]} com {mais_velha[1]} anos.")
else:
    print("\nNenhum dado foi informado.")
```

```python
''' Dado o dicion√°rio ao lado:
Imprima o total de epis√≥dios
Substitua a quantidade de epis√≥dios pelo
percentual (ep_anime/ep_total)
Imprima nome e percentual de cada
anime, em ordem decrescente de
percentual'''
animes = {
    "Naruto": 220,
    "Jujutsu Kaisen": 47,
    "Dragon Ball Z": 291,
    "Death Note": 37,
    "Fullmetal Alchemist": 64,
    "Evangelion": 26,
    "Berserk": 25,
    "Code Geass": 50,
    "Akame ga Kill!": 24,
    "Elfen Lied": 13
}

# 1. Imprima o total de epis√≥dios
total_episodios = sum(animes.values())
print(f"1. Total de epis√≥dios: {total_episodios}")

percentuais_animes = {}
for nome, episodios in animes.items():
    percentual = (episodios / total_episodios) * 100
    percentuais_animes[nome] = percentual

print("2. Dicion√°rio com Percentuais (para verifica√ß√£o):")
print({k: f"{v:.2f}%" for k, v in percentuais_animes.items()}) # Imprime formatado




animes_ordenados = sorted(percentuais_animes.items(), key=lambda item: item[1], reverse=True)

print("\n3. Animes por percentual (decrescente):")
for nome, percentual in animes_ordenados:
    print(f"- {nome}: {percentual:.2f}%")



```

```python
'''Dado o dicion√°rio ao lado:
Imprima o √°lbum mais antigo
imprima a banda que tem o maior
intervalo entre os √°lbuns
Imprima todos os √°lbuns do mais novo
para o mais antigo, dizendo tamb√©m o
nome da banda''''
bandas = {
    "Gangrena Gasosa": [
        {"√°lbum": "Gente Ruim S√≥ Manda Lembran√ßa Pra Quem N√£o Presta", "ano": 2019},
        {"√°lbum": "Se Deus √â 10, Satan√°s √â 666", "ano": 1999},
        {"√°lbum": "Welcome to the Macumba", "ano": 1993},
    ],
    "Rog√©rio Skylab": [
        {"√°lbum": "Nas Portas do Cu", "ano": 2023},
        {"√°lbum": "Cr√≠tica da Faculdade do Cu", "ano": 2021},
        {"√°lbum": "Cosmos", "ano": 2019},
    ],
    "Garotos Podres": [
        {"√°lbum": "Can√ß√µes para Ninar", "ano": 2003},
        {"√°lbum": "Com a Corda Toda", "ano": 1997},
        {"√°lbum": "Rock de Sub√∫rbio", "ano": 1995},
    ],
    "Massacration": [
        {"√°lbum": "Live Metal Espancation", "ano": 2017},
        {"√°lbum": "Good Blood Headbanguers", "ano": 2009},
        {"√°lbum": "Gates of Metal Fried Chicken of Death", "ano": 2005},
    ],
    "Raimundos": [
        {"√°lbum": "Cantigas de Roda", "ano": 2014},
        {"√°lbum": "Roda Viva", "ano": 2006},
        {"√°lbum": "√âramos 4", "ano": 2001},
    ]
}


album_mais_antigo = None
ano_mais_antigo = float('inf') # Come√ßa com um ano 'infinito'

for banda, albuns in bandas.items():
    for album in albuns:
        if album["ano"] < ano_mais_antigo:
            ano_mais_antigo = album["ano"]
            album_mais_antigo = f'{album["√°lbum"]} ({ano_mais_antigo}) da banda {banda}'

print(f"1. √Ålbum mais antigo: {album_mais_antigo}")


maior_intervalo = -1
banda_maior_intervalo = ""

for banda, albuns in bandas.items():

    anos = [album["ano"] for album in albuns]


    if anos: # Verifica se a lista n√£o est√° vazia
        ano_min = min(anos)
        ano_max = max(anos)
        intervalo = ano_max - ano_min

        if intervalo > maior_intervalo:
            maior_intervalo = intervalo
            banda_maior_intervalo = banda

print(f"2. Banda com o maior intervalo entre √°lbuns: {banda_maior_intervalo} ({maior_intervalo} anos)")



lista_todos_albuns = []

for banda, albuns in bandas.items():
    for album in albuns:

        lista_todos_albuns.append((album["ano"], album["√°lbum"], banda))


albuns_ordenados_por_ano = sorted(lista_todos_albuns, key=lambda item: item[0], reverse=True)

print("\n3. Todos os √°lbuns (do mais novo para o mais antigo):")
for ano, album, banda in albuns_ordenados_por_ano:
    print(f"- {album} ({ano}) - Banda: {banda}")


```
